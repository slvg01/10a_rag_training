```python
import configparser

# Read the configuration file
config = configparser.ConfigParser()
config.read('config.ini')

# Access the API token from the configuration file
huggingfacehub_api_token = config.get('huggingface', 'api_token')
# Set your API Key from OpenAI
openai_api_key = config.get('openai', 'api_key')

```

```python
from langchain_openai import OpenAI

# Define the LLM
llm = OpenAI(model_name="gpt-3.5-turbo-instruct", openai_api_key=openai_api_key)

# Predict the words following the text in question
question = 'dont worry , be sure that'
output = llm.invoke(question)

print(output)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

# Define an OpenAI chat model
llm = ChatOpenAI(temperature=0, openai_api_key=openai_api_key)		

# Create a chat prompt template
prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("human", "Respond to question: {question}")
    ]
)

# Insert a question into the template and call the model
full_prompt = prompt_template.format_messages(question='How can I retain learning?')
llm.invoke(full_prompt)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain



# Define the LLM
llm = HuggingFaceEndpoint(repo_id='tiiuae/falcon-7b-instruct', huggingfacehub_api_token=huggingfacehub_api_token)

question = "How does LangChain make LLM application development easier?"

# Create a prompt template from the template string
template = "You are an artificial intelligence assistant, answer the question. {question}"
prompt = PromptTemplate(template=template, input_variables=["question"])

# Create a chain to integrate the prompt template and LLM
llm_chain = LLMChain(llm=llm, prompt=prompt)

print(llm_chain.invoke(question))
```

```python
#use chatmessagememory
from langchain.memory import ChatMessageHistory
llm = HuggingFaceEndpoint(repo_id='tiiuae/falcon-7b-instruct', huggingfacehub_api_token=huggingfacehub_api_token)
history = ChatMessageHistory()
history.add_ai_message("Ask me anything about python programming")
history.add_user_message("what is list comprehension?")
llm_response = llm.invoke(history.messages)
print(llm_response)
history.add_user_message("describe with 15 words only")
llm_response= llm.invoke(history.messages)
print(llm_response)
```

```python
#with buffer memory    
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
llm = HuggingFaceEndpoint(repo_id='mistralai/Mistral-7B-Instruct-v0.2', huggingfacehub_api_token=huggingfacehub_api_token)
memory = ConversationBufferMemory(size=20 )
buffer_chain = ConversationChain(llm=llm, memory=memory, verbose=True)

buffer_chain.invoke("Write python code to draw a scatter plot with matplotlib library") 
buffer_chain.invoke("use seaborn library?")
buffer_chain.invoke("if i want an histogram")
buffer_chain.invoke("but i want the code for seaborn and in red")
                        
```

```python
#with conversation summary memory
from langchain.memory import ConversationSummaryMemory
from langchain.chains import ConversationChain
llm = HuggingFaceEndpoint(repo_id='tiiuae/falcon-7b-instruct', huggingfacehub_api_token=huggingfacehub_api_token)
memory = ConversationSummaryMemory(llm=llm)
summary_chain = ConversationChain(llm=llm, memory=memory, verbose=True)
summary_chain.invoke("what is the belgium surrealism movement?")
summary_chain.invoke("what is the difference between surrealism and dadaism?")
summary_chain.invoke("can you summerize those in one sentence")
summary_chain.invoke("can you summerize differences in 3 words?")   

```
